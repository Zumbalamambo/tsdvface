```python
import os
import random
import numpy as np
random.seed(0)

### Generate pairs.txt
DATA_DIR = '/home/hungnv/master/src/facenet/datasets/tsdv_face_mtcnnpy_160/'

### Make all matches
matches_list = []
for name in os.listdir(DATA_DIR):
    a = []
    num_pairs = 0
    for file in os.listdir(os.path.join(DATA_DIR, name)):
        a.append(file)
        
    if len(a) < 5:
        num_pairs = 3
    elif (len(a) >= 5) and (len(a) < 10):
        num_pairs = 5
    elif (len(a) >= 10) and (len(a) < 15):
        num_pairs = 10
    elif (len(a) >= 15) and (len(a) < 20):
        num_pairs = 20
    else:
        num_pairs = 26
    
    for i in range(num_pairs):
        while(1):
            name_1 = random.choice(a).split(".")[0]
            name_2 = random.choice(a).split(".")[0]
            if (name_1 != name_2):
                matches_list.append([name, name_1, name_2])
                break

random.shuffle(matches_list)
matches_list_final = matches_list[:300]


### Make all mismatches
mismatches_list = []
for idx, name in enumerate(os.listdir(DATA_DIR)):
        
    list_choice = []
    for i in range(10):
        remaining = os.listdir(DATA_DIR)
        del remaining[idx]    # deletes the file from the list, so that it is not chosen again
        
        while(1):
            other_dir = random.choice(remaining)
            
            if other_dir not in list_choice:
                a = []
                num_pairs = 0
                for file in os.listdir(os.path.join(DATA_DIR, name)):
                    a.append(file)
                    
                if len(a) < 5:
                    num_pairs = 3
                elif (len(a) >= 5) and (len(a) < 10):
                    num_pairs = 5
                elif (len(a) >= 10) and (len(a) < 15):
                    num_pairs = 10
                elif (len(a) >= 15) and (len(a) < 20):
                    num_pairs = 20
                else:
                    num_pairs = 30
                    
                for j in range(num_pairs):
                    
                    while(1):
                        file1 = random.choice(os.listdir(os.path.join(DATA_DIR, name)))
                        file2 = random.choice(os.listdir(os.path.join(DATA_DIR, other_dir)))

                        if file1[:-4] != file2[:-4]:
                            mismatches_list.append([name, file1[:-4], other_dir, file2[:-4]])  
                            break
                    

                list_choice.append(other_dir)
                
                break

random.shuffle(mismatches_list)
mismatches_list_final = mismatches_list[:1200]

print len(matches_list_final)
print len(mismatches_list_final)
 

final_list = matches_list_final + mismatches_list_final

random.shuffle(final_list)

# print total_list[:10]
print len(final_list)



with open('../data/pairs_tsdvface_20180920.txt', 'w') as f:
    for item in final_list:
        if len(item) == 3:
            f.write(item[0] + "\t" + item[1] + "\t" + item[2] + "\n")
        elif len(item) == 4:
            f.write(item[0] + "\t" + item[1] + "\t" + item[2] + "\t" + item[3] + "\n")
        else:
            None
```
